### Vamos iniciar tendo uma familiarização com as bibliotecas numpy, matplotlib e alguns algoritmos simples utilizados no machine learning sendo eles: Linear regression, Cost function e Gradient Descent.

```python
#VALORES DE TREINAMENTO
x = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55])
#y = np.array([25, 35, 45, 55, 65, 75, 85, 95, 105, 115])
y = np.array([24, 33, 45, 55, 60, 77, 85, 99, 108, 114])

plt.scatter (x,y)
plt.show()
```

```python
#IMPLEMENTAÇÃO DA COST FUNCTION
def cost_function(x,y,w,b):
    m=x.shape[0] #Pega o tamanho da primeira dimensão;
    cost = 0
    for i in range(m):
        cost += (w*x[i] + b - y[i])**2
    cost = cost/2*m
    return cost
        
```

```python
#IMPLEMENTAÇÃO DA FUNÇÃO GRADIENT DESCENT
def gradient_descent(x,y,w,b,iterations):
    m = x.shape[0]
    alpha = 0.001
    for j in range(iterations):
        dj_dw=0
        dj_db=0
        for i in range(m):
            dj_dw += (w*x[i] + b - y[i])*x[i]
            dj_db += (w*x[i] + b - y[i])
        w= w-(alpha/m)*dj_dw
        b = b- (alpha/m)*dj_db
    return(w,b)
```

```python
w,b = gradient_descent(x,y,1,1,100000)
print(w,b)
```

```python
y_t=w*x + b
plt.scatter (x,y)

plt.plot(x,y_t)
#plt.grid(True) Ativa a grade do gráfico
plt.show()

error =cost_function(x,y,w,b)
print (error)
```

### Vamos resolver esse problema usando o Tensorflow



```python
import tensorflow as tf  #O tensorrflow é uma biblioteca especializada em machine learning
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

#Esse é a arquitetura da nossa rede neural, o "Dense" é interpretado como uma camada
model = Sequential ([
    Dense(units=1, activation='linear', input_shape=(1,))]) #A última camada é a de saída ela leva a ativação que reflete o problema

#O parâmetro "units" da última camada deve corresponder ao número de saídas do seu modelo
#Nesse caso so temos uma única camada

#Agora importamos o nosso algoritmo de perdas, ou erros. Como uma cost function
from tensorflow.keras.losses import MeanSquaredError 

model.compile (optimizer=tf.keras.optimizers.Adam(learning_rate=0.003),
loss=MeanSquaredError()) #Atribuimos a função de erro desejada ao modelo

#.fit usa uma função semelhante, porém mais avançada que o gradient descent para treinar o modelo
#epochs é o número de iterações
model.fit(x,y,epochs=2000)
model.summary()

''''O algoritmo Adam otimiza a taxa de aprendizado do gradient descent.
Isto é: se é perceptível que podemos aumentar ou diminuir a taxa de aprendizado ele fará isso automaticamente
Adam: Adaptative Moment estimation'''
```

```python
predictions = model.predict(x)
plt.scatter (x,y)
plt.plot(x,predictions)
plt.show()
```

### Agora vamos resolver um problema clássico de classificação com uma quantidade de dados maior

#### Nesse conjunto de dados temos dois tipos de plantas clássificadas em 4 caractéristicas: comprimento de sepála, largura de sépala, comprimento de petála e largura de pétala. 

####  As duas plantas possíveis são:
#### A Iris-setosa


#### A Iris-versicolor




######  Note que as duas plantas são muito parecidas a olho nú.



//perguntar como ajustar a imagem diretamente no Markdown

```python
#Primeiro vamos carregar nosso conjunto de treinamento;
x_entrada = []
y_entrada =[]
x_teste =[]
y_teste =[]
class_to_number = {"Iris-setosa": 0, "Iris-versicolor" : 1} #Atribui 0 para Iris-setosa e 1 para Iris-versicolor
#Abre o arquivo para leitura
with open("iris.DATA", "r") as file:
    lines = file.readlines() #Lê todas as linhas do arquivo
    
#Agregando os valores as váriaveis:
for i in lines:
    parts = i.strip().split(',') #Dividir a linha nas vírgulas
    x_valores=([float(val) for val in parts[:4]])
    x_entrada.append(x_valores)
    y_valores = class_to_number[parts[4]]
    y_entrada.append(y_valores)
    
y_training=np.array(y_entrada, dtype=np.float32)
x_training = np.array(x_entrada, dtype=np.float32)

#Vamos importar também nossos dados de teste
with open("iris-teste.DATA", "r") as file:
    lines_teste = file.readlines()

for j in lines_teste:
    parts =j.strip().split(',')
    x_valores =[float(val) for val in parts[:4]]
    x_teste.append(x_valores)
    y_valores = class_to_number[parts[4]]
    y_teste.append(y_valores)
    
y_teste=np.array(y_teste, dtype=np.float32)    
x_teste= np.array(x_teste, dtype=np.float32)
```

```python
#Vamos observar nosso conjunto de dados:

print("Valores de treinanemnto X:\n",x_training)
print("Valores de treinamento Y:\n",y_training)
print("Valores de teste X:\n",x_teste)
print("Valores de teste Y:\n",y_teste)
```

```python
#Dimensões das váriaveis
print("Dimensões de X:", x_training.shape)
print("Dimensões de Y:", y_training.shape)
print("Dimensões de X_Teste:", x_teste.shape)
print("Dimensões de Y_teste:", y_teste.shape)
```

```python
#Agora vamos aplicar nosso algoritmo de aprendizado


#Esse é a arquitetura da nossa rede neural:
model_2 = Sequential ([
    Dense (units=4, activation= 'relu'), #Nas camadas ocultas de uma rede neural é utilizado comumente a ativação Relu; 
    Dense (units=15, activation= 'relu'),
    Dense (units=1, activation= 'sigmoid'), #A última camada é a de saída ela leva a ativação que reflete o problema
])
#O parâmetro "units" da primeira e última camada deve corresponder ao número de entradas e saídas do seu modelo respectivamente


#Agora importamos o nosso algoritmo de perdas, ou erros. Como uma cost function
from tensorflow.keras.losses import BinaryCrossentropy 

model_2.compile (optimizer=tf.keras.optimizers.Adam(learning_rate=0.003),
loss=BinaryCrossentropy()) #Atribuimos a função de erro desejada ao modelo

#.fit usa uma função semelhante, porém mais avançada que o gradient descent para treinar o modelo
#epochs é o número de iterações
model_2.fit(x_training,y_training,epochs=300)
model_2.summary()

''''O algoritmo Adam otimiza a taxa de aprendizado do gradient descent.
Isto é: se é perceptível que podemos aumentar ou diminuir a taxa de aprendizado ele fará isso automaticamente
Adam: Adaptative Moment estimation'''
```

```python
#Agora vamos usar nossos valores de entrada de teste para ver as respostas obtidas pelo modelo e comparar com as esperadas

y_1=model_2.predict(x_teste)
m = y_1.shape[0]

for i in range (m):
    if (y_1[i] >= 0.5):
        print("Iris-Versicolor")
    else:
        print("Iris-setosa")
        
print("\nComparando com os resultados esperados dos nossos dados:\n")

t=y_teste.shape[0]

for i in range(t):
    if (y_teste[i] >= 0.5):
        print("Iris-Versicolor")
    else:
        print("Iris-setosa")
```
